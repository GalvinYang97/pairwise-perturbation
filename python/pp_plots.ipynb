```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import genfromtxt
```

# Plot function: 

```python
def plot_(tucker_dt, tucker_pp):
    dt_iter = tucker_dt[:,1]
    dt_diffnorm = tucker_dt[:,2]
    dt_ppupdate = tucker_dt[:,4]
    dt_diffv = tucker_dt[:,5]
    dt_dtime = tucker_dt[:,6]
    pp_iter = tucker_pp[:,1]
    pp_diffnorm = tucker_pp[:,2]
    pp_ppupdate = tucker_pp[:,4]
    pp_diffv = tucker_pp[:,5]
    pp_dtime = tucker_pp[:,6]
    pp_restart = tucker_pp[:,7]

    plt.figure(figsize=(14,7))
    plt.plot(dt_dtime, dt_diffnorm, marker='s',linewidth=2,label='dimension tree')
    plt.plot(pp_dtime, pp_diffnorm, marker='o',linewidth=2,label='pairwise perturbation')
#     plt.plot(dt_dtime, dt_diffnorm, linewidth=2,label='dimension tree')
#     plt.plot(pp_dtime, pp_diffnorm, linewidth=2,label='pairwise perturbation')

    # get the start incice
#     pp_start = 0
#     for i in range(len(pp_ppupdate)):
#         if(pp_ppupdate[i]!=0):
#             pp_start = i
#             break
#     labels[pp_start] = 'PP starts'

#     for label, x, y in zip(labels[pp_start:pp_start+1], pp_dtime[pp_start:pp_start+1], pp_diffnorm[pp_start:pp_start+1]):
#         plt.annotate(
#             label,
#             xy=(x, y), xytext=(40, 20),
#             textcoords='offset points', ha='right', va='bottom',
#             bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.2),
#             arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))
    labels = ['PP starts'.format(i) for i in pp_iter]
    for i in range(len(pp_restart)-1):
        if(pp_ppupdate[i+1]!=0 and pp_ppupdate[i]==0):
            label = labels[i]
            x = pp_dtime[i]
            y = pp_diffnorm[i]
    #     for label, x, y in zip(labels[restart:restart+1], pp_dtime[restart:restart+1], pp_diffnorm[restart:restart+1]):
            plt.annotate(
                label,
                xy=(x, y), xytext=(-20, -40),
                textcoords='offset points', ha='right', va='bottom',
                bbox=dict(boxstyle='round,pad=0.5', fc='blue', alpha=0.2),
                arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))
    labels = ['DT starts'.format(i) for i in pp_iter]
    for i in range(len(pp_restart)-1):
        if(pp_ppupdate[i+1]==0 and pp_ppupdate[i]!=0):
            label = labels[i]
            x = pp_dtime[i]
            y = pp_diffnorm[i]
    #     for label, x, y in zip(labels[restart:restart+1], pp_dtime[restart:restart+1], pp_diffnorm[restart:restart+1]):
            plt.annotate(
                label,
                xy=(x, y), xytext=(40, 20),
                textcoords='offset points', ha='right', va='bottom',
                bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.2),
                arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))
        
        
    plt.legend()
    plt.xlabel('Time(s)')
    plt.ylabel('step error of the core tensor')
    plt.yscale("log")
    plt.show()
```

# case with Tucker decomposition of a possion opearator
dimension of the tensor: 4 <br>
size in each dim: 40  <br>
core tensor size: 10  <br>
steperror: ||core_tensor|| - ||core tensor of previous step|| / ||core_tensor||   <br>
restart condition : exist ||dA||/||A|| > 1e-2  <br>

```python
tucker_dt = genfromtxt('tucker_dt_4_40_10_ps.csv', delimiter=',')[1:]
tucker_pp = genfromtxt('tucker_pp_4_40_10_ps.csv', delimiter=',')[1:]
plot_(tucker_dt, tucker_pp)
```
